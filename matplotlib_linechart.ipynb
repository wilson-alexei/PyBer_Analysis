```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month. (5.1.3)
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot 
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend. (5.1.4)
plt.plot(x_axis,y_axis,label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Month")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Add a grid
plt.grid()
# Create and format the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis,color="green",marker="X",linewidth=2,label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Months')
ax.set_ylabel('Fare($)')
ax.grid()
```
